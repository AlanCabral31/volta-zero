#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('SINASC_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(10, 6))

sns.countplot(x='APGAR1', data=df)
plt.title('Frequência - APGAR1')
plt.show()

sns.countplot(x='APGAR5', data=df)
plt.title('Frequência - APGAR5')
plt.show()

sns.countplot(x='SEXO', data=df)
plt.title('Frequência - SEXO')
plt.show()

sns.countplot(x='GESTACAO', data=df)
plt.title('Frequência - GESTACAO')
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='ESTCIVMAE', data=df)
plt.title('Frequência - ESTCIVMAE')
plt.xticks(rotation=45)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure', figsize=(10, 6))

sns.barplot(x='CONSULTAS', y='APGAR5', data=df)
plt.title('APGAR5 médio por número de CONSULTAS')
plt.xticks(rotation=45)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(12, 10))

fig, axes = plt.subplots(2, 2)

sns.countplot(ax=axes[0, 0], x='SEXO', data=df)
axes[0, 0].set_title('Frequência - SEXO')

sns.countplot(ax=axes[0, 1], x='GESTACAO', data=df)
axes[0, 1].set_title('Frequência - GESTACAO')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1, 0], x='ESTCIVMAE', data=df)
axes[1, 0].set_title('Frequência - ESTCIVMAE')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1, 1], x='CONSULTAS', data=df)
axes[1, 1].set_title('Frequência - CONSULTAS')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
