####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('SINASC_RO_2019.csv')

sns.boxplot(x=df['IDADEMAE'])
plt.title('Boxplot - Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(x=df['PESO'])
plt.title('Boxplot - Peso do Bebê')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
df['IDADEMAE'].plot.hist(bins=20)
plt.title('Histograma - Idade da Mãe')
plt.xlabel('Idade')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
df['PESO'].plot.hist(bins=20)
plt.title('Histograma - Peso do Bebê')
plt.xlabel('Peso (gramas)')
plt.show()
```
